```python
import pandas as pd
import glob
import os
```

```python
folder_path = r"C:\Users\ASUS_\p2_model\data"
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))

print("Total CSV files:", len(csv_files))
```

```python
merged_file_path = r"C:\Users\ASUS_\p2_model\merged_dataset.csv"

first_file = True

for file in csv_files:
    print("Processing:", file)
    
    chunk = pd.read_csv(file)
    
    if first_file:
        chunk.to_csv(merged_file_path, index=False, mode='w')
        first_file = False
    else:
        chunk.to_csv(merged_file_path, index=False, mode='a', header=False)

print("✅ All files merged successfully!")
```
